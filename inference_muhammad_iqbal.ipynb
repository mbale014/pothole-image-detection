Pada bagian ini akan dilakukan prediksi dengan model yg sudah di training sebelumnya. Gambar yg di prediksi diluar dataset train-val-set.

```python
#Import library
import numpy as np
from PIL import Image
from tensorflow.keras.models import load_model
```

```python
# model_link = 'https://drive.google.com/file/d/1CEK-vVdbIhcUqJqBXPB8o8dGvsKMFnE9/view?usp=drive_link'

#Mount to googel drive API
from google.colab import drive
drive.mount('/content/drive')

#Load model
model = load_model('/content/drive/MyDrive/FTDS_hacktiv8/GC7/best_model.h5')
```

```python
# Create a function for test a new image

def predict_image(img_path):
    img = Image.open(img_path)
    img = img.resize((224, 224))
    img_array = np.asarray(img)
    img_array = img_array / 255.0  # same as training
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)[0][0]
    class_name = 'Pothole' if prediction > 0.5 else 'Normal'
    print(f"Predicted class: {class_name} ({prediction:.2f})")
```
